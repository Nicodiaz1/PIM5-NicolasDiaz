# Cargar Datos

Este notebook carga el dataset de ejemplo no productivo para el pipeline de MLOps.  
En un entorno productivo, los datos provienen del DWH o Data Lake de la empresa.  
Aquí usamos un archivo local como sustituto para desarrollo y pruebas.


## 1. Importar librerías


```python
import pandas as pd
import os
```

## 2. Definir ruta del archivo

El dataset es un archivo `.xlsx` de ejemplo no productivo.  
Se guarda la ruta de forma relativa para que funcione en cualquier equipo.


```python
RUTA_DATOS = '../../Base_de_datos.xlsx'
RUTA_CSV   = '../../Base_de_datos.csv'

print(f"Ruta datos: {RUTA_DATOS}")
```

## 3. Cargar el dataset


```python
df = pd.read_excel(RUTA_DATOS)
print(f"Dataset cargado correctamente.")
print(f"Filas: {df.shape[0]} | Columnas: {df.shape[1]}")
```

## 4. Vista previa del dataset


```python
df.head()
```

## 5. Información general del dataset


```python
df.info()
```

## 6. Guardar el dataset como CSV

Convertimos el archivo a `.csv` para que el resto del pipeline lo consuma de forma estándar.


```python
df.to_csv(RUTA_CSV, index=False)

print(f"Archivo guardado en: {RUTA_CSV}")
```
